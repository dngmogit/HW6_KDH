```python
import csv
import matplotlib.pyplot as plt

def read(data):
    num = [0]*6
    i=0
    
    for i in range(6):
        num[i]=int(next(data)[1])
    return num
        
    
def graph(title, nums):
    plt.title(title)
    plt.xlabel("1 ~ 6 numbers")
    plt.ylabel("trials")
    plt.bar([1,2,3,4,5,6],nums)
    plt.show()
    

def main():
    f = open('result.csv','r',encoding='cp949')
    data = csv.reader(f)
    
    one = next(data)[0]
    one_result = read(data)
    
    two = next(data)[0]
    two_result = read(data)
    
    three = next(data)[0]
    three_result = read(data)
    
    four = next(data)[0]
    four_result = read(data)
    
    graph(one,one_result)
    graph(two,two_result)
    graph(three,three_result)
    graph(four,four_result)
    f.close()
    print('  q2.txt   .')
    
if __name__ == "__main__":
    main()
    
```
