```python
import csv
import matplotlib.pyplot as plt

def read(data):
    a = [0]*2
    next(data)
    jeju = next(data)
    a[0] = float(jeju[-3])
    a[1] = float(jeju[-2])
    result = a[0]/a[1]
    return result 
    
def main():
    one = open('2008.csv','r',encoding='cp949')
    two = open('2013.csv','r',encoding='cp949')
    three = open('2018.csv','r',encoding='cp949')
    four = open('2022.csv','r',encoding='cp949')
    
    one_r = csv.reader(one)
    two_r = csv.reader(two)
    three_r = csv.reader(three)
    four_r = csv.reader(four)
    
    
    years = [2008,2013,2018,2022]
    result = [read(one_r),read(two_r),read(three_r),read(four_r)]
    
    plt.title("Gender Ratio in Jeju Island")
    plt.xlabel("Years")
    plt.ylabel("Ratio (Man/Woman)")
    plt.bar(years,result)
    plt.show()
    
    print("2008 year : ",result[0])
    print("2013 year : ",result[1])
    print("2018 year : ",result[2])
    print("2022 year : ",result[3])
    print('결과 분석은 q3.txt 파일을 참고해주시기 바랍니다.')
    
    one.close()
    two.close()
    three.close()
    four.close()
    
    
if __name__ == "__main__":
    main()
    
```
